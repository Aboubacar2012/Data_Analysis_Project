#### FEATURES

- id: The ID of the movie (clear/unique identifier)
- title: The official title of the movie.
- tagline: The of the movie.
- release_date: Theatrical Realease Date of the movie.
- genres: Genres associated with movie.
- belongs_to_collection: Gives information on the movie series/franchise the particular to.
- original_language: The language in which the movie was originally shot in.
- budget_musd: The budget of the movie in million dollars
- revenue_musd: The total revenue of the movie in million dollars.
- production_companies:Production companies involved with the making of the movie;
- production_countries: Countries where the movie was shot in/produced in
- vote_count:the number rating of the movie.
- popularity: The popularity Score assigned by TMDB
- runtime: The runtime of the movie in minutes.
- overview : A brief blurb of the movie
- spoken_languages:Spoken languages in the film
- poster_path: The URL of the poster image. 
- cast : (Main) Actors appearing in the movie
- cast_size:number of Actors appearing in the movie.
- director : Director of the movie.
- crew_size:Size of the film crew(incl.director, excl. actors)

######The best and the worst movies...(Part 1)

```python
from IPython.display import HTML
```

```python
df_best=df[['poster_path','title','budget_musd','revenue_musd','vote_count','vote_average','popularity']].copy()
df_best
```

```python
df_best['revenue_musd']=df.revenue_musd.sub(df.budget_musd)
df_best['return']=df.revenue_musd.div(df.budget_musd)
```

```python
df_best
```

```python
#Rename colummns headers
df_best.columns=['','Title','Buget','Revenue','Vote','Average Rating','Popularity','ROI']
df_best.set_index('Title', inplace=True)
```

```python
df_best
```

```python
df_best.iloc[0,0]
```

```python
subset=df_best.iloc[:5,:2]
subset
```

```python
HTML(subset.to_html(escape=False))
```

#####Sorting

```python
# Sort by Average Rating
df_best.sort_values(by='Average Rating', ascending=False)
```

```python
# Sort by ROI
df_best.sort_values(by='ROI', ascending=False)
```

```python
# Sort by ROI with buget >= 5 millions (Method 1)
df_best.loc[df_best.Buget>=5].sort_values(by='ROI', ascending=False)
```

```python
#Method 2
df_best[df_best['Buget']>=5].sort_values(by='ROI', ascending=False)
```

##### Fill NaN by O

```python
df_best.isnull().sum()
```

```python
df_best.Buget.fillna(0,inplace=True)
df_best.Vote.fillna(0, inplace=True)
```

```python
df_best.info()
```

```python
# Best and worst Movies
# Let's define a function
def best_worst(n, by, ascending=False, min_bud=0, min_votes=0):
  df2=df_best.loc[(df_best.Buget>=min_bud)&(df_best.Vote>=min_votes),
                  ['', by]].sort_values(by=by, ascending=ascending).head(n).copy()
  return HTML(df2.to_html(escape=False))
```
